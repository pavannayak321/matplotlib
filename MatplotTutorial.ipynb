```
import matplotlib.pyplot as plt
from matplotlib import style


style.use('ggplot')


x1 = [1,2,3,4,5]
y1 = [23,34,54,12,1]

plt.xlabel('Day')
plt.ylabel('Score')

plt.plot(x1,y1)
plt.title('Score On Match Day')

plt.show()
```

```
plt.subplot(2,1,1)
plt.plot([1,2,3,4],[21,23,43,21])
plt.title('First Subplot')



plt.subplot(2,1,2)
plt.scatter([1,2,3,4],[21,23,21,22])
plt.title('2Nd plot')


plt.suptitle('Subplots')
plt.show()
```

```
#creating the subplots


fig , ax = plt.subplots(nrows=2,ncols=2,figsize=(12,12))
ax[0,0].plot([1,2,3,4],[21,12,14,43])
ax[1,0].scatter([1,2,4,8],[12,34,56,13],color='g')

plt.show()
```

```
import matplotlib.pyplot as plt


#bar graph
v1 = ['2011','2012','2013','2014']
#style.use('ggplot')
company_A = [2000,1000,2300,500]
company_B = [300,2000,4300,5000]
company_C = [540,2300,2000,1000]
company_D = [100,200,5600,3200]


plt.bar(v1,company_A, label = 'COMPANY_A',align='center',color = 'g')

plt.bar(v1,company_B, label = 'COMPANY_B',align='center',color = 'r')

plt.bar(v1,company_C, label = 'COMPANY_C',align='center',color = 'b')


plt.bar(v1,company_D, label = 'COMPANY_D',align='center',color = 'k')

plt.grid(True,color = 'G')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Bar')
plt.legend()
plt.show()
```
