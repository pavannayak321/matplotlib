```python
#pie chart cab be best used in case of stock prediction/representation
%matplotlib inline
import matplotlib.pyplot as plt

profit_week = [122,233,332,123,324]
source      = ['oppo','nokia','mi','htc','samsung']
color      = ['c','m','r','b','k']


plt.pie(profit_week,labels=source,colors=color,shadow= True,explode=(0,0.1,0,0,0),autopct='%1.4f%%')
plt.show()
```
