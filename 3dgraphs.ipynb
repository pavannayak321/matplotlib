```python
%matplotlib inline


from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

#creating figure object
fig = plt.figure()

#creating subplot  object and adding axis to it
ax = fig.add_subplot(111,projection='3d')

xs = [1,2,3,4,5,6]
ys = [12,13,12,14,21,12]
zs = [0,0,0,0,0,0]

ax.scatter(xs,ys,zs,color='r',marker='*',zdir='z',depthshade=True)
ax.set_xlabel('x Axis') 
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axes')
plt.show()
```

```python
%matplotlib inline


from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

#creating figure object
fig = plt.figure()

#creating subplot  object and adding axis to it
ax = fig.add_subplot(111,projection='3d')

xs = [1,2,3,4,5,6]
ys = [12,13,12,14,21,12]
zs = [0,0,0,0,0,0]

ax.plot_surface(xs,ys)
ax.set_xlabel('x Axis') 
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axes')
plt.show()
```
