```python
%matplotlib inline
import matplotlib.pyplot as plt
x = [1,2,3,3,4]
y= [12,3,4,5,2] 
plt.plot(x,y)
plt.ylabel('some numbers')
plt.xlabel('xlabel')
```

```python
#labelling the lines
%matplotlib inline
x= [1,2,3,4]
y= [1,4,9,16]
plt.plot(x,y,label="First Line")

plt.plot(y,x,label="Second line")
plt.legend()
plt.show()
```

```python
#Bar graphs are used to compare things between different groups or to track changes over time.
x = ['p','q','r','s']
y = [9,7,6,5]
plt.xlabel('xlabel')
plt.ylabel('ylabel')

plt.bar(x,y,label='bargraph')
plt.legend()
plt.show()
```
