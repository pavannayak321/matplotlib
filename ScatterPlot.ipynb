'''Next up, we cover scatter plots! The idea of scatter plots is usually to compare two variables, or three if you are plotting in 3 dimensions, looking for correlation or groups.
'''

```python
%matplotlib inline
import matplotlib.pyplot as plt
temperature = [14,16,11,15,18,22,19,25,23,18,22,17]
ice_cream_sale= [214,234,234,213,239,321,212,121,232,344,232,231]

plt.xlabel('temparature')
plt.ylabel('ice_cream_sale')
plt.title('Scatter Plot')


plt.scatter(temperature,ice_cream_sale,label='ice cream sale',color='g',marker='^',s=39)
plt.legend()
plt.show()
```
